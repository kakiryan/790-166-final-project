```python
# networkx clusteirng not supported for the multigraph type (a multigraph has multiple connections between the same pairs of nodes) 
# --> transform the multigraph into weighted digraph 
```

```python
new_graph = nx.Graph()
for u, v in graph.edges():
    if new_graph.has_edge(u, v):
        new_graph[u][v]['weight'] += 1
    else:
        new_graph.add_edge(u, v, weight=1)
```

```python
new_graph.edges()
```

```python
len(new_graph) # same amount of nodes
```

```python
len(new_graph.edges()) # less edges
```

```python
new_graph['n0'] # now we have weights instead of 32 separate edges
```

```python
# https://networkx.org/documentation/latest/reference/algorithms/generated/networkx.algorithms.cluster.clustering.html
clusters = nx.clustering(new_graph, weight='weight')
clusters
```

```python
adj_matrix = nx.to_numpy_matrix(new_graph) # convert to adj matrix
adj_matrix
```

```python
kmeans = KMeans(n_clusters=12)
kmeans.fit(adj_matrix)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
from sklearn.cluster import SpectralClustering
sc = SpectralClustering(12, affinity='precomputed', n_init=10, assign_labels='discretize')
results = sc.fit(adj_matrix)
```

```python
results.labels_
```

```python
from sklearn.metrics import normalized_mutual_info_score

normalized_mutual_info_score(kmeans.labels_, results.labels_)
```
